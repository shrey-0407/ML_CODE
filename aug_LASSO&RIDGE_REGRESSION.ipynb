```python
df1=df.drop(['Address','Date','YearBuilt','Lattitude','Longtitude','Postcode'],axis=1)
df1
```

```python
df1.isna().sum()
```

```python
col_to_zero=['Bedroom2','Bathroom','Car','Propertycount','Distance']
df1[col_to_zero]=df1[col_to_zero].fillna(0)
df1.isna().sum()
```

```python
df1['Landsize']=df1['Landsize'].fillna(df1.Landsize.mean())
df1['BuildingArea']=df1['BuildingArea'].fillna(df1.BuildingArea.mean())
```

```python
df1.isna().sum()
```

```python
df1.dropna(inplace=True)
df1.isna().sum()
```

```python
df1.head()
```

### ONE-HOT ENCODING

```python
df1=pd.get_dummies(df1,drop_first=True)
df1
```

```python
x=df1.drop('Price',axis=1)
y=df1['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=linear_model.Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```
